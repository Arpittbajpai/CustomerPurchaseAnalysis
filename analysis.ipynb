```python
import pandas as pd
from google.colab import files
import os
```

```python
uploaded = files.upload()
df = pd.read_csv("customer_analysis.csv")

```

```python
df.head()

```

```python
def category(x):
    if x > 30000:
        return "premium"
    else:
        return "regular"

df["customer_category"] = df["Purchase_Amount"].apply(category)
```

```python
df
```

```python
def customer_age_wise(x):
    if x <= 30:
        return "young"
    elif x <= 40:
        return "middle aged"
    else:
        return "senior"

df["customer_age_wise"] = df["Age"].apply(customer_age_wise)
```
